```python
import pandas as pd
import time
from datetime import datetime as dt
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from collections import OrderedDict

symptomdata = pd.read_csv('2021_country_daily_2021_GB_daily_symptoms_dataset.csv')
casedata = pd.read_csv('ltla_2021-07-25.csv')

casedata = casedata.sort_values(by=['areaName','date'], ascending=(True, True))
uniquenames = list()
for index, row in symptomdata.iterrows():
    for name in list(OrderedDict.fromkeys(casedata['areaName'].tolist())):
        if row['sub_region_1'] == name:
            uniquenames.append(row['sub_region_1'])
            
uniquenames = list(OrderedDict.fromkeys(uniquenames))

uniquenames
```
